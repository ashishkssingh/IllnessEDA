```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('darkgrid')
```

```python
df = pd.read_csv('/kaggle/input/toy-dataset/toy_dataset.csv')
```

```python
income_stats = [df['Income'].mean(), df['Income'].mode()[0], df['Income'].median()]
age_stats = [df['Age'].mean(), df['Age'].mode()[0], df['Age'].median()]

index = ['Mean', 'Mode', 'Median']

mean_mode_median = pd.DataFrame({'Income' : income_stats, 'Age' : age_stats}, index=index)

print('Mean, Mode and Median for Income and Age')
mean_mode_median
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Gender', data=df)
```

```python
plt.figure(figsize=(8,5))
sns.distplot(df[df['Gender'] == "Male"]['Income'], hist=False, kde_kws={"shade": True},label= "Male Income" )
sns.distplot(df[df['Gender'] == "Female"]['Income'], hist=False, kde_kws={"shade": True},label = "Female Income")
```

```python
plt.figure(figsize=(8,5))
plt.xticks(rotation=90)
sns.countplot(x='City', data=df)
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Illness', data=df)
```

```python
plt.figure(figsize=(10,8))
sns.countplot(x="City", hue="Illness", data=df)
plt.show()
```
